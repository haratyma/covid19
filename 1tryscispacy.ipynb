```python
import numpy as np 
import pandas as pd 
import spacy
from pprint import pprint
```

```python
import scispacy
from scispacy.umls_linking import UmlsEntityLinker 

nlp = spacy.load("en_core_sci_lg")
linker = UmlsEntityLinker(resolve_abbreviations=True)
nlp.add_pipe(linker)
```

```python
df_meta=pd.read_csv('/hackathon/covid-19/data/metadata.csv')
```

```python
text = df_meta['abstract'][0]
```

```python
fdoc = nlp(text)
```

```python
spacy.displacy.render(fdoc, style='ent', jupyter=True)
pprint([(ent.text, ent.label_) for ent in fdoc.ents])
```

```python
def entity_links(text,concept_id=0):
    doc = nlp(text)
    spacy.displacy.render(doc, style='ent', jupyter=True)
    pprint([(ent.text, ent.label_) for ent in doc.ents])
    print("Entity name", doc.ents[concept_id])
    for umls_ent in (doc.ents[concept_id])._.umls_ents:
        print(linker.umls.cui_to_entity[umls_ent[0]])
    
```

```python
entity_links("Known risk factors for the coronavirus disease are: older age, male gender, diabetes and leukemia. Female patiens and children are less susceptible.",1)
```
